## Recommendation code

the structure is pretty simple: if person liked/favorited some stories, he will probably make a transaction in category connected to the story category

```python
df_liked_fav['categories_for_recommend'] = df_liked_fav['category']
df_liked_fav['last_day_for_recommendation'] = df_liked_fav['date_time']
df_liked_fav['categories_for_recommend'] = df_liked_fav['categories_for_recommend'].astype(object)
for index, row in tqdm(df_liked_fav.iterrows()):
    if df_liked_fav.loc[index, 'category'] in list(dict_of_categories.keys()):
#         print (df_liked_fav.loc[index, 'category'] )
#         print(dict_of_categories[df_liked_fav.loc[index, 'category']])
        df_liked_fav.at[index, 'categories_for_recommend'] = list(dict_of_categories[df_liked_fav.loc[index, 'category']])
        df_liked_fav.loc[index, 'last_day_for_recommendation'] = datetime.datetime.fromtimestamp(int(time.mktime(df_liked_fav.loc[index, 'date_time'].timetuple())) + 2592000)
    else:
        df_liked_fav.at[index, 'categories_for_recommend'] = np.nan
        df_liked_fav.loc[index, 'last_day_for_recommendation'] = np.nan
```

Let's test it on data that we have

```python
without_nan_df_liked_fav = df_liked_fav.dropna()
without_nan_df_liked_fav
```

```python
from collections import Counter
correct_rec = 0
incorrect_rec = 0
transactions_correct_rec = 0
for index, row in tqdm(without_nan_df_liked_fav.iterrows()):
    buffer_df = df6[df6.party_rk == without_nan_df_liked_fav.loc[index, 'party_rk']]
    buffer_df = buffer_df[buffer_df.transaction_dttm > without_nan_df_liked_fav.loc[index, 'date_time']]
    buffer_df = buffer_df[buffer_df.transaction_dttm < without_nan_df_liked_fav.loc[index, 'last_day_for_recommendation']]
    counter_dict = Counter(list(buffer_df.category))
    for rec_cat in without_nan_df_liked_fav.loc[index, 'categories_for_recommend']:
        if rec_cat in list(buffer_df.category):
            correct_rec+=1
        if rec_cat in list(counter_dict.keys()):
            transactions_correct_rec += counter_dict[rec_cat]
        else:
            incorrect_rec += 1
```

```python
print(correct_rec, incorrect_rec, transactions_correct_rec)
```
